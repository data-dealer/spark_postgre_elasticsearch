```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder \
    .appName("PySparkPostgresWrite") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.6.0") \
    .getOrCreate()
```

```python
import requests
import json
import subprocess
import os
import time
```

```python
df = spark.read.parquet("s3a://datalake/fake_data_1M").where('batch <= 30')
```

```python
start_time = time.time()
(
    df.write.format("jdbc")
    .option("url", "jdbc:postgresql://postgres-db:5432/postgres")
    .option("driver", "org.postgresql.Driver")
    .option("dbtable", "fake_1_3")
    .option("user", "postgres")
    .option("password", "postgres")
    .save()
)
total_time = (time.time() - start_time)/60
print('total write time:',  total_time)
```
